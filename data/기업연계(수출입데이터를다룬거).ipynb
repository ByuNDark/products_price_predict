```python
import pandas as pd
import numpy as np
```

```python
토마토 = pd.read_csv('토마토수출입.csv', encoding='cp949')
포도 = pd.read_csv('포도수출입.csv', encoding='cp949')
```

```python
포도
```

```python
포도.columns = ['연도','월','수출(중량)','수출(금액)','수입(중량)','수입(금액)']
```

```python
포도
```

```python
중간단계 = 포도.drop(포도[포도['월'] == '합'].index)
```

```python
중간단계 = 중간단계.reset_index(drop=True)
```

```python
중간단계
```

```python
for x in range(59):
    중간단계['연도'].iloc[x] = int(중간단계['연도'].iloc[x])
    중간단계['월'].iloc[x] = int(중간단계['월'].iloc[x])
    중간단계['수출(중량)'].iloc[x] = float(중간단계['수출(중량)'].iloc[x].replace(',', ''))
    중간단계['수출(금액)'].iloc[x] = float(중간단계['수출(금액)'].iloc[x].replace(',', ''))
    중간단계['수입(중량)'].iloc[x] = float(중간단계['수입(중량)'].iloc[x].replace(',', ''))
    중간단계['수입(금액)'].iloc[x] = float(중간단계['수입(금액)'].iloc[x].replace(',', ''))
```

```python
중간단계
```

```python
A = [];B = [];C = [];D = []
for x in range(59):
    if (x == 1) | (x == 49):
        for i in range(29):
            A.append(중간단계['수출(중량)'].iloc[x] / 29)
            B.append(중간단계['수출(금액)'].iloc[x] / 29)
            C.append(중간단계['수입(중량)'].iloc[x] / 29)
            D.append(중간단계['수입(금액)'].iloc[x] / 29)
    elif x in [13, 25, 37]:
        for i in range(28):
            A.append(중간단계['수출(중량)'].iloc[x] / 28)
            B.append(중간단계['수출(금액)'].iloc[x] / 28)
            C.append(중간단계['수입(중량)'].iloc[x] / 28)
            D.append(중간단계['수입(금액)'].iloc[x] / 28)
    elif x in [3,5,8,10, 15,17,20,22, 27,29,32,34, 39,41,44,46, 51,53,56]:
        for i in range(30):
            A.append(중간단계['수출(중량)'].iloc[x] / 30)
            B.append(중간단계['수출(금액)'].iloc[x] / 30)
            C.append(중간단계['수입(중량)'].iloc[x] / 30)
            D.append(중간단계['수입(금액)'].iloc[x] / 30)
    elif x == 58:
        for i in range(5):
            A.append(중간단계['수출(중량)'].iloc[x] / 30)
            B.append(중간단계['수출(금액)'].iloc[x] / 30)
            C.append(중간단계['수입(중량)'].iloc[x] / 30)
            D.append(중간단계['수입(금액)'].iloc[x] / 30)
    else:
        for i in range(31):
            A.append(중간단계['수출(중량)'].iloc[x] / 31)
            B.append(중간단계['수출(금액)'].iloc[x] / 31)
            C.append(중간단계['수입(중량)'].iloc[x] / 31)
            D.append(중간단계['수입(금액)'].iloc[x] / 31)
```

```python
테스트 = pd.DataFrame({'수출(중량)':A,
                    '수출(금액)':B,
                    '수입(중량)':C,
                    '수입(금액)':D})
테스트
```

```python
테스트.to_csv("포도_수출입.csv", encoding='utf-8-sig')
```
