```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.5f}'.format
```

```python
농산물전체 = pd.read_csv('/content/drive/MyDrive/대기실/train.csv')
배추 = 농산물전체[['date', '요일', '배추_거래량(kg)', '배추_가격(원/kg)']]
```

```python
# 1주 전 가격들의 평균내기(월~일, 0으로 된 날 개수 세는 코드와 그에 따른 평균 내는 코드로)
# index 기준 10~16, 17~23, ...
# index 기준 0~9는 농넷에서 찾아야 할 듯 합니다. so 2015년 12월 28부터 2015년 12월 31일까지 4일간의 데이터 찾아보자.
A = [];B = []
for x in range(10):
    A.append(0)
    B.append(0)

for y in range(246):
    s = 0;t = 0
    for z in range(7):
        s += 배추['배추_거래량(kg)'].iloc[(y * 7) + z + 3]
        t += 배추['배추_가격(원/kg)'].iloc[(y * 7) + z + 3]
    v_1 = (배추['배추_거래량(kg)'].iloc[(y * 7) + 3:(y * 7) + 10] == 0).sum()
    v_2 = (배추['배추_가격(원/kg)'].iloc[(y * 7) + 3:(y * 7) + 10] == 0).sum()
    for w in range(7):
        A.append(s / (7 - v_1))
        B.append(t / (7 - v_2))
for z in range(7):
    s += 배추['배추_거래량(kg)'].iloc[1725 + z]
    t += 배추['배추_가격(원/kg)'].iloc[1725 + z]
v_1 = (배추['배추_거래량(kg)'].iloc[1725:1732] == 0).sum()
v_2 = (배추['배추_가격(원/kg)'].iloc[1725:1732] == 0).sum()
A.append(s / (7 - v_1))
B.append(t / (7 - v_2))
Adf = pd.DataFrame(A, columns = ['배추_거래량의_1주전_평균'])
Bdf = pd.DataFrame(B, columns = ['배추_가격의_1주전_평균'])
새로운배추 = pd.concat([배추, Adf, Bdf], axis=1)
새로운배추
```

```python
# 2주 전 가격들의 평균내기(월~일, 0으로 된 날 개수 세는 코드와 그에 따른 평균 내는 코드로)
# index 기준 10~23, 24~37, ...
# index 기준 0~16는 농넷에서 찾아야 할 듯 합니다. so 2015년 12월 21부터 2015년 12월 31일까지 11일간의 데이터 찾아보자.
C = [];D = []
for x in range(17):
    C.append(0)
    D.append(0)

for y in range(245):
    s = 0;t = 0
    for z in range(14):
        s += 배추['배추_거래량(kg)'].iloc[(y * 7) + z + 3]
        t += 배추['배추_가격(원/kg)'].iloc[(y * 7) + z + 3]
    v_1 = (배추['배추_거래량(kg)'].iloc[(y * 7) + 3:(y * 7) + 17] == 0).sum()
    v_2 = (배추['배추_가격(원/kg)'].iloc[(y * 7) + 3:(y * 7) + 17] == 0).sum()
    for w in range(7):
        C.append(s / (14 - v_1))
        D.append(t / (14 - v_2))
for z in range(14):
    s += 배추['배추_거래량(kg)'].iloc[1718 + z]
    t += 배추['배추_가격(원/kg)'].iloc[1718 + z]
v_1 = (배추['배추_거래량(kg)'].iloc[1718:1732] == 0).sum()
v_2 = (배추['배추_가격(원/kg)'].iloc[1718:1732] == 0).sum()
C.append(s / (14 - v_1))
D.append(t / (14 - v_2))
Cdf = pd.DataFrame(C, columns = ['배추_거래량의_2주전_평균'])
Ddf = pd.DataFrame(D, columns = ['배추_가격의_2주전_평균'])
새로운배추 = pd.concat([배추, Cdf, Ddf], axis=1)
새로운배추
```
