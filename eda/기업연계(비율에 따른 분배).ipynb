```python
import random
import pandas as pd
import numpy as np
import random
```

```python
기온_1 = [];강수량_1 = [];평균운량_1 = []
기온_2 = [];강수량_2 = [];평균운량_2 = []
기온_3 = [];강수량_3 = [];평균운량_3 = []
for x in range(100):
    기온_1.append(random.randint(10, 21))
    기온_2.append(random.randint(15, 28))
    기온_3.append(random.randint(20, 33))
    강수량_1.append(random.randint(0, 5))
    강수량_2.append(random.randint(0, 25))
    강수량_3.append(random.randint(0, 15))
    평균운량_1.append(random.randint(0, 10))
    평균운량_2.append(random.randint(0, 10))
    평균운량_3.append(random.randint(0, 10))

df = pd.DataFrame({'지역1기온':기온_1,
                  '지역2기온':기온_2,
                  '지역3기온':기온_3,
                  '지역1강수량':강수량_1,
                  '지역2강수량':강수량_2,
                  '지역3강수량':강수량_3,
                  '지역1평균운량':평균운량_1,
                  '지역2평균운량':평균운량_2,
                  '지역3평균운량':평균운량_3})
df
```

```python
# 지역1 재배량 63%, 지역2 재배량 21%, 지역3 재배량 16%라고 가정한다면?
final = df.copy()
final['기온의_적절한_평균'] = df['지역1기온'] * (63/100) + df['지역2기온'] * (21/100) + df['지역3기온'] * (16/100)
final['강수량의_적절한_평균'] = df['지역1강수량'] * (63/100) + df['지역2강수량'] * (21/100) + df['지역3강수량'] * (16/100)
final['평균운량의_적절한_평균'] = df['지역1평균운량'] * (63/100) + df['지역2평균운량'] * (21/100) + df['지역3평균운량'] * (16/100)
final = final.drop(['지역1기온', '지역2기온', '지역3기온', '지역1강수량', '지역2강수량', '지역3강수량', '지역1평균운량', '지역2평균운량', '지역3평균운량'], axis=1)
final
```

```python
# def : df에서의 지역1기온을 최저기온, 지역3기온을 최고기온으로 하자. 일교차 구하는 법
df['일교차'] = abs(df['지역1기온'] - df['지역3기온'])
df
```
