```python
# 1주 전 가격들의 평균내기(월~일, 0으로 된 날 개수 세는 코드와 그에 따른 평균 내는 코드로)
# index 기준 10~16, 17~23, ...
# index 기준 0~9는 걍 그 날 가격으로 합니다.
# this is 1733row version.
A = [];B = [];a = (len(대파) - 10) // 7;b = (len(대파) - 10) % 7
for x in range(10):
    A.append(대파['대파_거래량(kg)'].iloc[x])
    B.append(대파['대파_가격(원/kg)'].iloc[x])
for y in range(a):
    s = 0;t = 0
    for z in range(7):
        s += 대파['대파_거래량(kg)'].iloc[(y * 7) + z + 3]
        t += 대파['대파_가격(원/kg)'].iloc[(y * 7) + z + 3]
    v_1 = (대파['대파_거래량(kg)'].iloc[(y * 7) + 3:(y * 7) + 10] == 0).sum()
    v_2 = (대파['대파_가격(원/kg)'].iloc[(y * 7) + 3:(y * 7) + 10] == 0).sum()
    for w in range(7):
        A.append(s / (7 - v_1))
        B.append(t / (7 - v_2))
for z in range(7):
    s += 대파['대파_거래량(kg)'].iloc[1725 + z]
    t += 대파['대파_가격(원/kg)'].iloc[1725 + z]
v_1 = (대파['대파_거래량(kg)'].iloc[1725:1732] == 0).sum()
v_2 = (대파['대파_가격(원/kg)'].iloc[1725:1732] == 0).sum()
for w in range(b):
    A.append(s / (7 - v_1))
    B.append(t / (7 - v_2))
Adf = pd.DataFrame(A, columns = ['대파_거래량의_1주전_평균'])
Bdf = pd.DataFrame(B, columns = ['대파_가격의_1주전_평균'])

# 2주 전 가격들의 평균내기(월~일, 0으로 된 날 개수 세는 코드와 그에 따른 평균 내는 코드로)
# index 기준 10~23, 24~37, ...
# index 기준 0~16는 걍 그 날 가격으로 합니다.
C = [];D = []
for x in range(17):
    C.append(대파['대파_거래량(kg)'].iloc[x])
    D.append(대파['대파_가격(원/kg)'].iloc[x])
for y in range(a - 1):
    p = 0;q = 0
    for z in range(14):
        p += 대파['대파_거래량(kg)'].iloc[(y * 7) + z + 3]
        q += 대파['대파_가격(원/kg)'].iloc[(y * 7) + z + 3]
    v_1 = (대파['대파_거래량(kg)'].iloc[(y * 7) + 3:(y * 7) + 17] == 0).sum()
    v_2 = (대파['대파_가격(원/kg)'].iloc[(y * 7) + 3:(y * 7) + 17] == 0).sum()
    for w in range(7):
        C.append(p / (14 - v_1))
        D.append(q / (14 - v_2))
for z in range(14):
    p += 대파['대파_거래량(kg)'].iloc[1718 + z]
    q += 대파['대파_가격(원/kg)'].iloc[1718 + z]
v_1 = (대파['대파_거래량(kg)'].iloc[1718:1732] == 0).sum()
v_2 = (대파['대파_가격(원/kg)'].iloc[1718:1732] == 0).sum()
for w in range(b):
    C.append(p / (14 - v_1))
    D.append(q / (14 - v_2))
Cdf = pd.DataFrame(C, columns = ['대파_거래량의_2주전_평균'])
Ddf = pd.DataFrame(D, columns = ['대파_가격의_2주전_평균'])

새로운대파 = pd.concat([대파, Adf, Bdf, Cdf, Ddf, season[:1733]], axis=1)
```
